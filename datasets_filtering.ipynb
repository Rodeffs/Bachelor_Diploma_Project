# Работа с датасетами

## Библиотеки

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

## Параметры

Корневая папка со всеми датасетами

```python
root_dir = Path(r"./datasets")
```

Закрывающийся магазин. Указать id в виде строки

```python
closing_store = "10"
```

Считать ли алкогольные товары отдельно. Это нужно, так как не у всех организаций есть лицензия на продажу алкоголя. В таком случае, нужно искать наилучшее распределение среди магазинов, которые состоят в той же организации, что и закрывающийся

```python
separate_alcohol = False
```

Минимальный срок хранения распределяемого товара. Это обусловлено тем, что процесс закрытия магазина не быстрый, обычно занимает не меньше месяца времени, потому скоропортящиеся товары в любом случае не успеют быть распределены.

Значение переменной 0 означает, что срок годности любой

```python
spoilage_threshold = 30
```

Не все товары измеряются поштучно, некоторые измеряются по весу, другие - по длине. В таком случае, нужно определить, сколько килограмм или метров содержит условно одна штука товара

```python
kilos_in_one_piece = 1.0
meters_in_one_piece = 1.0
```

Стоит ли искать магазины с тем же куратором или организацией, что и закрывающийся магазин. Это нужно, так как контакты между разными организациями или кураторами добавляют лишние издержки, проще когда всё происходит в рамках одной компании или куратора.

```python
same_curator = False
same_organization = False
```

## Подгрузка датасетов

```python
assortment = pd.read_csv(root_dir / "Ассортимент" / "Assortment.csv", sep=";", dtype={"№Магазина" : "str"})  # т.к. под номером может быть РЦ
assortment.head()
```

```python
leftovers = pd.read_csv(root_dir / "Остатки" / "Stocks.csv", sep=";", dtype={"номер магазина" : "str"})
leftovers.head()
```

```python
facings = pd.read_csv(root_dir / "Планограмма" / "Export_facings.csv", sep=";", dtype={"STORENUMBER" : "str"})
facings.head()
```

```python
stores = pd.read_excel(root_dir / "SPR_Магазины_Сеть_РД.xlsx", dtype={"Магазин" : "str"})
stores.head()
```

```python
nomenclature = pd.read_excel(root_dir / "SPR_Номенклатура_NEW.xlsx")
nomenclature.head()
```

```python
sales = pd.read_csv(root_dir / "Продажи" / "SPR_Среднедневные_продажи.csv", sep=",", dtype={"Магазин" : "str"})
sales.head()
```

Переименование одинаковых колонок

```python
col_vol = "Объём товара в закрывающемся магазине (V)"
col_maxcap = "Максимальная вместимость товара в магазине (M)"
col_item = "Код товара"
col_store = "Магазин"
col_status = "Ассортимент Статус"
col_remain = "Остаток товара в магазине (R)"
col_sales = "Продажи товара в магазине (S)"
col_measure = "Единица измерения"
col_left = "остаток_робот"
```

```python
assortment = assortment.rename(columns={"№Магазина" : col_store, "Номер SKU" : col_item})
leftovers = leftovers.rename(columns={"номер магазина" : col_store, "код товара" : col_item})
facings = facings.rename(columns={"STORENUMBER" : col_store, "STOCKCODE" : col_item})
stores = stores.rename(columns={"Магазин" : col_store})
nomenclature = nomenclature.rename(columns={"Код" : col_item, "ОсновнаяЕдиницаИзмерения" : col_measure})
sales = sales.rename(columns={"Код товара" : col_item, "Магазин" : col_store, "Продажи, ед. изм." : col_sales})
```

```python
leftovers[col_left] = leftovers[col_left].replace(r"\s+", "", regex=True)  # убрать пробелы в числах
leftovers[col_left] = leftovers[col_left].replace(",", ".", regex=True)  # т.к. по стандарту разделение через .
leftovers[col_left] = pd.to_numeric(leftovers[col_left])
leftovers[col_left] = leftovers[col_left].fillna(0)
leftovers.loc[leftovers[col_left] < 0, col_left] = 0  # если остаток отрицательный, то полагается, что товара нет на складе
leftovers.head()
```

## Предварительная работа с датасетами

Товары закрывающегося магазина

```python
items = leftovers[leftovers[col_store] == closing_store]
items = items.rename(columns={col_left : col_vol})
```

Номенклатура и срок годности товаров в закрывающемся магазине

```python
items = items.merge(nomenclature, left_on=col_item, right_on=col_item, how="left").filter(items=[col_item, "Уровень3", col_vol, "Срок хранения товара", col_measure])
items["Срок хранения товара"] = items["Срок хранения товара"].fillna(0)  # на случай, если срок годности не указан, ставим 0 и считаем непортящимся
items.head()
```

Фильтрация по сроку хранения товара. Если срок хранения товара равен 0, то товар считается непортящимся

```python
if spoilage_threshold > 0:
    items = items[(items["Срок хранения товара"] == 0) | (items["Срок хранения товара"] >= spoilage_threshold)]
items.head()
```

Магазины, где распределяемые товары состоят в продаже

```python
potential_stores = assortment[assortment[col_store] != closing_store]
items = items.merge(potential_stores, left_on=col_item, right_on=col_item).filter(items=[col_item, "Уровень3", col_store, col_status, col_vol, col_measure])
items = items[items[col_status] == "Введен"]
items.head()
```

Определяем куратора и организацию этих магазинов

```python
items = items.merge(stores, left_on=col_store, right_on=col_store).filter(items=[col_item, "Уровень3", col_store, "Куратор", "Организация", col_vol, col_measure])
items.head()
```

Куратор и организация закрывающегося магазина

```python
closing_curator, closing_organization = stores[stores[col_store] == closing_store].filter(items=["Куратор", "Организация"]).iloc[0].values
print("Куратор:", closing_curator)
print("Организация:", closing_organization)
```

Теперь добавляем среднедневные продажи товаров по магазинам

```python
items = items.merge(sales, left_on=[col_item, col_store], right_on=[col_item, col_store]).filter(items=[col_item, "Уровень3", col_store, "Куратор", "Организация", col_vol, col_sales, col_measure])
items.head()
```

Найдём размерность полок в магазинах. Очевидно, если дублируется магазин и код товара, то полок несколько и их нужно сложить

```python
items = items.merge(facings, left_on=[col_item, col_store], right_on=[col_item, col_store]).filter(items=[col_item, "Уровень3", col_store, "Куратор", "Организация", col_vol, col_sales, col_measure, "XFACINGS", "YFACINGS", "ZFACINGS"])
items[["XFACINGS", "YFACINGS", "ZFACINGS"]] = items[["XFACINGS", "YFACINGS", "ZFACINGS"]].replace(0, 1)  # 0 быть не может физически, как минимум 1 предмет влезет
items[col_maxcap] = items["XFACINGS"] * items["YFACINGS"] * items["ZFACINGS"]  # получаем максимальную вместимость
items = items.drop(columns=["XFACINGS", "YFACINGS", "ZFACINGS"])

# Сумма дублей
duplicates = items[items.duplicated(subset=[col_item, "Уровень3", col_store, "Куратор", "Организация", col_vol, col_sales, col_measure], keep=False)]
unduplicated = duplicates.groupby([col_item, "Уровень3", col_store, "Куратор", "Организация", col_vol, col_sales, col_measure], as_index=False)[col_maxcap].sum()

# Удаляем дубликаты из исходной
items = items.drop_duplicates(subset=[col_item, col_store], keep=False)

# Вместо них добавим уже просуммированные значения
items = pd.concat([items, unduplicated])
items.head()
```

Остатки товара в каждом магазине

```python
items = items.merge(leftovers, left_on=[col_item, col_store], right_on=[col_item, col_store]).filter(items=[col_item, "Уровень3", col_store, "Куратор", "Организация", col_vol, col_sales, col_maxcap, col_left, col_measure])
items = items.rename(columns={col_left : col_remain})
items.head()
```

Смотрим на основную единицу измерения и если это кг или м, то делим их на количество кг или м в одной штуке

```python
items.loc[items[col_measure] == "кг", [col_vol, col_maxcap, col_remain]] /= kilos_in_one_piece
items.loc[items[col_measure] == "м", [col_vol, col_maxcap, col_remain]] /= meters_in_one_piece
```

Нецелые данные о количестве товара округляем

```python
items = items.round({col_vol : 0, col_maxcap : 0, col_remain : 0})
```

Делим на алкогольную и безалкогольную продукцию

```python
alcohol_items = None

if separate_alcohol:
    alcohol_items = items[
        (items["Уровень3"] == "0601 Напитки безалкогольные") | 
        (items["Уровень3"] == "0602 Водка и Настойки") |
        (items["Уровень3"] == "0603 Крепкие спиртные напитки") |
        (items["Уровень3"] == "0604 Игристые вина") |
        (items["Уровень3"] == "0605 Вермут") |
        (items["Уровень3"] == "0606 Вино") |
        (items["Уровень3"] == "0607 Слабоалкогольные напитки")]

    items = items[
        (items["Уровень3"] != "0601 Напитки безалкогольные") & 
        (items["Уровень3"] != "0602 Водка и Настойки") &
        (items["Уровень3"] != "0603 Крепкие спиртные напитки") &
        (items["Уровень3"] != "0604 Игристые вина") &
        (items["Уровень3"] != "0605 Вермут") &
        (items["Уровень3"] != "0606 Вино") &
        (items["Уровень3"] != "0607 Слабоалкогольные напитки")]
```

Если это требуется, оставляем только магазины с тем же куратором или организацией

```python
if same_curator and same_organization:
    items = items[(items["Куратор"] == closing_curator) | (items["Организация"] == closing_organization)]

elif same_curator:
    items = items[items["Куратор"] == closing_curator]

elif same_organization:
    items = items[items["Организация"] == closing_organization]

items = items.filter(items=[col_item, col_store, col_vol, col_sales, col_maxcap, col_remain])  # оставляем только нужные колонки
```

Для алкогольной продукции всегда оставляем магазины только в пределах этой же организации

```python
if alcohol_items is not None:
    alcohol_items = alcohol_items[alcohol_items["Организация"] == closing_organization]
    alcohol_items = alcohol_items.filter(items=[col_item, col_store, col_vol, col_sales, col_maxcap, col_remain])  # оставляем только нужные колонки
```
